```python
%%sql
SELECT *
FROM access_to_basic_services
LIMIT 10;
```

## 1. Number of observations in the dataset, Minimum and Maximum Time period, Number of Countries involved:

```python
%%sql
SELECT
    COUNT(*) AS Number_of_observations,
    MIN(Time_period) AS Minimum_time_period,
    MAX(Time_period) AS Maximum_time_period,
    COUNT(DISTINCT Country_name) AS Number_of_countries,
    AVG(Pct_managed_drinking_water_services) AS AVG_managed_drinking_water_services
FROM
    Access_to_Basic_services;
```

## 2. Creating a Summary Statistic Report for Access to Drinking Water Services of each Sub-Region:

There are 18 Geographical Sub-Regions represented in this dataset with Number of countries calculated for each sub-region: 

There are _18 Geographical Sub-Regions_ represented in this dataset with _Number of countries_ calculated for each sub-region: 

```python
%%sql
SELECT
    Region,
    Sub_region,
    MIN(Pct_managed_drinking_water_services) AS Minimum_Pct_managed_drinking_water_services,
    MAX(Pct_managed_drinking_water_services) AS Maximum_Pct_managed_drinking_water_services,
    AVG(Pct_managed_drinking_water_services) AS Average_Pct_managed_drinking_water_services,
    COUNT(DISTINCT Country_name) AS Number_of_countries,
    SUM(Est_gdp_in_billions) AS EST_total_gdp_in_billions
FROM
    Access_to_basic_services
GROUP BY Region, Sub_region
ORDER BY SUM(Est_gdp_in_billions) ASC;
```

## 3. Which Regions and Sub-regions had the Lowest Access to Drinking Water services in _2020?_

```python
%%sql
SELECT
    Region,
    Sub_region,
    MIN(Pct_managed_drinking_water_services) AS Min_Pct_managed_drinking_water_services,
    MAX(Pct_managed_drinking_water_services) AS Max_Pct_managed_drinking_water_services,
    AVG(Pct_managed_drinking_water_services) AS Avg_Pct_managed_drinking_water_services,
    COUNT(DISTINCT Country_name) AS Countries,
    SUM(Est_gdp_in_billions) AS Total_gdp_in_billions
FROM
    Access_to_basic_services
WHERE
    Time_period = 2020
AND Pct_managed_drinking_water_services < 60
GROUP BY Region, Sub_region
ORDER BY AVG(Pct_managed_drinking_water_services) ASC; 
```

## 4. What is the Average Access to Sanitation services in Sub-Saharan Africa in _2020?_

```python
%%sql
SELECT
    Region,
    Sub_region,
    ROUND(AVG(Pct_managed_sanitation_services), 2) AS Average_Access_to_Sanitation
FROM
    Access_to_basic_services
WHERE 
    Time_period = 2020
AND 
    Region = 'Sub-Saharan Africa'
GROUP BY Region, Sub_region;
```

## 5. Which countries require the MOST assistance from the United Nations due to Low Access to the Basic Services?

```python
%%sql
SELECT
    Country_name,
    ROUND(AVG(Pct_managed_drinking_water_services)) AS Average_Access_to_Water,
    ROUND(AVG(Pct_managed_sanitation_services)) AS Average_Access_to_Sanitation
FROM
    Access_to_Basic_services
GROUP BY Country_name
HAVING
    AVG(Pct_managed_drinking_water_services) < 50
    AND AVG(Pct_managed_sanitation_services) < 50;
```

## 7. Grouping West African Countries into Regional Economic Communities

There are 3 MAIN Regional Economic Communities for West African Countries:
- SADC : South African Development Community
- UMA : Union du Maghreb Arabe 
- ECOWAS : Economic Community of West African States

```python
%%sql
SELECT 
    CASE 
        WHEN Country_name IN ("Angola", "Botswana", "Comoros", "Democratic Republic of Congo",
                              "Eswatini", "Lesotho", "Madagascar", "Malawi", "Mauritius", 
                              "Namibia", "Seychelles", "South Africa", "United Republic of Tanzania",
                              "Zambia", "Zimbabwe")
            THEN "SADC"
        WHEN Country_name IN ("Algeria", "Libya", "Mauritania", "Tunisia")
            THEN "UMA"
        WHEN Country_name IN ("Benin", "Burkina Faso", "Cabo Verde", "Cote d'Ivoire", "Gambia",
                              "Ghana", "Guinea", "Guinea-Bissau", "Liberia", "Mali", "Niger",
                              "Nigeria", "Senegal", "Sierra Leone", "Togo")
            THEN "ECOWAS"
        ELSE "Not Classified"
    END AS Regional_Economic_Community,
    Country_name,
    Pct_managed_drinking_water_services
FROM 
    Access_to_Basic_services
WHERE Country_name IN ("Angola", "Botswana", "Comoros", "Democratic Republic of Congo",
                       "Eswatini", "Lesotho", "Madagascar", "Malawi", "Mauritius", 
                       "Namibia", "Seychelles", "South Africa", "United Republic of Tanzania",
                       "Zambia", "Zimbabwe", "Algeria", "Libya", "Mauritania", "Tunisia",
                       "Benin", "Burkina Faso", "Cabo Verde", "Cote d'Ivoire", "Gambia",
                       "Ghana", "Guinea", "Guinea-Bissau", "Liberia", "Mali", "Niger",
                       "Nigeria", "Senegal", "Sierra Leone", "Togo")
AND Time_period = 2020;
```

### 8. Evaluating GDP Per Capita and Income Group based on the Poverty Line

In 2020, the Poverty Line Index was 2.50:
- Any index below the Poverty Line is termed as _Low Income Country_
- Any index above the Poverty Line is termed as _High Income Country_
- Any index equal to the Poverty Line is termed as _Medium Income Country_

Each Country is Grouped into either of the three income groups above, based on 
their __GDP per Capita_ 

- GDP Per Capita = GDP in Billions / Total Population
- GDP Per Capita per Day = GDP Per Capita / 365.25


-The GDP Per Capita Per Day is compared to the Poverty Line Index

```python
%%sql
SELECT DISTINCT Country_name,
    Time_period,
    Est_population_in_millions,
    Est_gdp_in_billions,
    ROUND((Est_gdp_in_billions / Est_population_in_millions * 1000), 2)  AS GDP_per_capita,
    ROUND((Est_gdp_in_billions / Est_population_in_millions * 1000) / 365.25, 2) AS GDP_per_capita_per_day,
    CASE 
        WHEN ROUND((Est_gdp_in_billions / Est_population_in_millions * 1000) / 365.25, 2) < 2.50 
            THEN "Low Income"
        WHEN ROUND((Est_gdp_in_billions / Est_population_in_millions * 1000) / 365.25, 2) > 2.50
            THEN "High Income"
        ELSE "Medium Income"
    END AS Income_Group
FROM
    Access_to_basic_services
WHERE 
    Time_period = 2020
    AND Est_gdp_in_billions IS NOT NULL;
```
